# 1. ENEL data

```python
enel_model = DataImputation()
```

## 1.1. Set parameters

### a. set_data_parameters()

```python
enel_model.set_data_parameters(data_folder_path = "/content/drive/Shared drives/electrical_resarch-data_imputation/data/enel/",
                               data_file_name = "codensa_97783.parquet",
                               df_type = "enel",
                               training_percentage = 0.7)
```

### b. set_autoencoder_parameters()

```python
enel_model.set_autoencoder_parameters(n_input_parameters_AE = 24,
                                      n_hidden_layers_AE = 10,
                                      n_neuron_hidden_layers_AE = 15,
                                      latent_dim_AE = 5,
                                      epochs_AE = 300,
                                      batch_size_AE = 10)
```

### c. set_ANN_hidden_variables_parameters()

```python
enel_model.set_ANN_hidden_varibles_parameters(n_hidden_layers_ANN_HV = 7,
                                              n_neuron_hidden_layers_ANN_HV = 15,
                                              epochs_HV = 100,
                                              batch_size_HV = 10)
```

### d.set_ANN_daily_load_parameters()

```python
enel_model.set_ANN_daily_load_parameters( n_hidden_layers_DL = 7,
                                         n_neuron_hidden_layers_DL = 15,
                                          epochs_DL = 100,
                                          batch_size_DL = 10)
```

## 1.2. Load data from the source file

```python
enel_model.load_data_in_models()
```

## 1.3. Training models

```python
enel_model.training_AE()
```

```python
enel_model.training_ANN_HV()
```

```python
enel_model.training_ANN_DL()
```

## 1.4. get_test_dataframe()

```python
enel_test_dataframe = enel_model.get_test_dataframe()

print(f"Dataframe shape is {enel_test_dataframe.shape}")
enel_test_dataframe.head()
```

```python
enel_day_select = enel_test_dataframe.iloc[:1]
print("Values for day select:")
print(f" - day_of_year: {enel_day_select['day_of_year'].values}")
print(f" - type_of_day_x: {enel_day_select[['type_of_day_1','type_of_day_2','type_of_day_3']].values}")
```

```python
enel_profile_day_select = enel_day_select[[str(hour) for hour in range(24)]]
```

## 1.5. predict profile

```python
enel_profile_predict = enel_model.predict_profile(day_of_year = 52, type_of_day = 1, decimal_randomness = 5)
```

## 1.6. Graphic comparison of synthetic and real profile

```python
X_axis_values = [str(hour) for hour in range(24)]
Y_axis_values_real = enel_profile_day_select.T
Y_axis_values_synthetic = enel_profile_predict.values.T


# Plotting both the curves simultaneously
plt.plot(X_axis_values, Y_axis_values_real, color='r', label='real')
plt.plot(X_axis_values, Y_axis_values_synthetic, color='g', label='synthetic')

# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Hours")
plt.ylabel("Wh")
plt.title("comparison of synthetic and real profile - ENEL")

# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
plt.grid()
# To load the display window
plt.show()
plt.close()
```
